# Problem 2

Now It is going to be a bit more difficult. Neither state is going to be balanced. You are asked to find a state that by changing only two weights (one on the left plate for one on the right), you can stay in a balanced situation. 

Note: we can see the gaps in the plates like 0 kg weights.

```python
inputs_problem2 = [[2,1,1,0,0,1],
                   [3,0,0,1,0,0],
                   [0,0,2,3,0,0],
                   [1,2,3,0,2,0]] # <--
```

In this example, the solution  is [1,2,3,0,2,0], because we can change the 2 kg weight of the left for a 0 kg weight of the right. In that way we will have 4 kg in the two sides.
